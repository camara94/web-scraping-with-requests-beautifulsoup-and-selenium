```python
# qui recupère de la page detail de chaque page
def get_url_tunisianet(urls=[], url=None, n=0):
    url_articles = []
    if n > 2:
        for page in range(2,n):
            url = f'{url}={page}'
            response = requests.get(url)
            page_contents = response.text
            if response.status_code != 200:
                raise Exception('Failed to load page {}'.format(items_url))
            doc = BeautifulSoup(page_contents, "html.parser")
            for item in doc.find_all("a", {'class': "thumbnail product-thumbnail first-img"}):
                url_articles.append(item['href'])
   
    for page in urls:
        url = page
        response = requests.get(url)
        page_contents = response.text
        if response.status_code != 200:
            raise Exception('Failed to load page {}'.format(items_url))
        doc = BeautifulSoup(page_contents, "html.parser")
        for item in doc.find_all("a", {'class': "thumbnail product-thumbnail first-img"}):
            url_articles.append(item['href'])
        
    return url_articles
```

```python
# La fonction qui les urls chacune des images
def get_article(items_url):
    images_articles = []
    # télécharger la page
    response = requests.get(items_url)
    # vérifier le succès de réponse
    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(items_url))
    # Parser la réponse à l'aide de beaufifulSoup
    doc = BeautifulSoup(response.text, 'html.parser')
    for i, img in enumerate(doc.find_all('a', {'class': 'thumb-container'})):
        if i>= 1 and len(doc.find_all('a', {'class': 'thumb-container'})) > 1:
            images_articles.append(img['data-image'])
    return images_articles
```

```python
import random
import urllib.request
import os

# La fonction de telechargement des images
def enregistrer_images(urls, doc):
    
    os.makedirs(os.path.join('images', doc))
    for i, url in enumerate(urls):
        try:
            fullname =  "images/" + doc + "/" + str((i+1))+".jpg"
            urllib.request.urlretrieve(url,fullname)
        except:
            pass
```

```python
# la fonction principale qui les lance toutes les autres 
def tunisianet_scraping_images(urls, url, n, dir_name):
    images = []
    url_article = get_url_tunisianet(urls, url, n)
    for uri in url_article:
        for u in get_article(uri):
            images.append(u)
            
    enregistrer_images(images, dir_name)
    return len(f'{len(images)} images')
```
