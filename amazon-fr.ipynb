# Web Scrapping avec Selenium

```python
#Si nous devons exécuter Selenium en mode sans entête ; c'est sans voir l'interface graphique en cours d'exécution, utilisez le code ci-dessous
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver import Chrome
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys

import time
import json
import os 
```

```python
def get_item_detail(url):
    driver.get(url)
    
    try:
        if len(driver.find_element_by_css_selector("#productTitle").text)>=1:
            titre = driver.find_element_by_css_selector("#productTitle").text
    except:
        titre = None
        
    try:
        if len(driver.find_element_by_css_selector("#productDescription p").text)>=1:
            description = driver.find_element_by_css_selector("#productDescription p").text
    except:
        description = None
    
    try:
        if len(driver.find_element_by_css_selector("#feature-bullets .a-list-item").text)>=1:
            apropos = driver.find_element_by_css_selector("#feature-bullets .a-list-item").text
    except:
        apropos = None


    try:
        # if len(driver.find_element_by_css_selector("#priceAndPrime").text)>=1:
        prix = driver.find_element_by_css_selector("#priceAndPrime").text
    except:
        prix = None
            
    try:
        if len(driver.find_element_by_css_selector(".imgTagWrapper img").get_attribute("src"))>=1:
            image_url = driver.find_element_by_css_selector(".imgTagWrapper img").get_attribute("src")
            print(driver.find_element_by_css_selector(".imgTagWrapper img").get_attribute("href"))
    except:
        image_url = None

    output_item = {"titre": titre, 'description':description,  "prix": prix, "url": image_url}
    return  output_item  

```

```python

def get_items(url):
    driver.get(url)
    # sleep for two second then scroll to the bottom of the page
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(4) #sleep between loadings
    produit_url = ""

    items = driver.find_elements_by_css_selector(".s-card-container")
    output = []

    for i in items:
        
        try:
            if len(i.find_element_by_css_selector(".a-link-normal").get_attribute("href"))>=1:
                produit_url = i.find_element_by_css_selector(".a-link-normal").get_attribute("href")
        except:
            produit_url = None
            
    
        output.append(produit_url)    
    return output



#elements-offer-price-normal__price
driverPath_1 = 'C:\Program Files\Google\Chrome\Application\chromedriver.exe'
serviceConfig = Service(driverPath_1)
optionConfig = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=serviceConfig, options=optionConfig)
    

page = 1
search = "ordinateur" #since we would like to search for Inflight Items
all_items = {}
url = 'https://www.amazon.fr/s?k=ordinateur&qid=1646836680&ref=sr_pg_1'
driver.get(url)
while True: 
    print(page)
    print(url)
    wait = WebDriverWait(driver, 10) 
    all_items[page] = get_items(url)

    #  by clicking on the last element of pagination == >
    search = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '.s-pagination-next')))
   # if it is active click
    try:
        
        search.send_keys(Keys.ENTER)
        url = driver.current_url
        time.sleep(3)
        page +=1
    except:
        pass
    if page == 3:
        break
    
# save all the items to a json file


driver.close()
```

```python
urls = []
for i in range(1,3):
    for u in all_items[i]:
        urls.append(u)
```
